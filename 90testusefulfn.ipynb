```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the sigmoid function
def abs_sigmoid(x,bias = 10):
  return 1 / (1 + math.exp(-abs(x)  + bias)*0.8)

# Generate x values
x = np.linspace(-20, 20, 400)

# Calculate the sigmoid values
y = [abs_sigmoid(i) for i in x]

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the exponential function
def abs_exponential(x, base = 5):
  return base ** abs(x)

# Generate x values
x = np.linspace(-50, 50, 1000)

# Calculate the exponential values
y = [abs_exponential(i) for i in x]

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
