```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the sigmoid function
def abs_sigmoid(x,bias = 10):
  return 1 / (1 + math.exp(-abs(x)  + bias)*0.8)

# Generate x values
x = np.linspace(-20, 20, 400)

# Calculate the sigmoid values
y = [abs_sigmoid(i) for i in x]

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def smooth_transition(x, n, k):
    """Smooth transition function from linear to exponential behavior."""
    sigmoid = 1 / (1 + np.exp(-k * (x - n)))
    linear_part = x
    exponential_part = np.exp(x - n) + n - 1
    return linear_part * (1 - sigmoid) + exponential_part * sigmoid

# Parameters
n = 4  # Transition point
k = 10000  # Smoothness of the transition

# Generate x values
x_values = np.linspace(3,5, 500)

# Calculate y values
y_values = smooth_transition(x_values, n, k)

# Plot the function
plt.plot(x_values, y_values, label=f"Transition at n={n}, k={k}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Smooth Transition from Linear to Exponential')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the exponential function
def abs_exponential(x, base = 5):
  return base ** abs(x)

# Generate x values
x = np.linspace(-50, 50, 1000)

# Calculate the exponential values
y = [abs_exponential(i) for i in x]

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
